```python
import datetime
import pandas as pd
import numpy as np
import pyodbc
pyodbc.drivers()
```

```python
raw_df = pd.read_csv('BD-Imp.csv',sep=';')
```

```python
raw_df.head()
```

```python
header = ['FCT_YR_NUM','FCT_MT_NUM','FCT_CD','FCT_DT',
          'CNTRY_CD','CMPTTR_NM','MTRL_NM','VLM_QTY',
          'VLM_UNT_CD','CIF_PRC_AMNT','CIF_PRC_CRRNCY_CD',
          'BDGT_CIF_PRC_AMNT','BDGT_CIF_PRC_CRRNCY_CD','PPLTN_DT']
```

```python
cln_df = pd.DataFrame(columns=header)
```

```python
cln_df['FCT_DT']=raw_df['Fecha']
cln_df['CNTRY_CD']=raw_df['País']
cln_df['CMPTTR_NM']=raw_df['Competidor']
cln_df['MTRL_NM']=raw_df['Descriptor']
cln_df['VLM_QTY']=raw_df['TM']
cln_df['CIF_PRC_AMNT']=raw_df['Precio CIF ($/TM)']
cln_df['BDGT_CIF_PRC_AMNT']=raw_df['PPTO Precio CIF ($/TM)']
```

```python
cln_df.replace('',None)
```

```python
cln_df['VLM_QTY']=cln_df['VLM_QTY'].astype(float)
```

```python
cln_df['FCT_DT']=cln_df['FCT_DT'].apply(lambda x: datetime.datetime.strptime(x, '%d/%m/%y'))
cln_df['FCT_YR_NUM']=cln_df['FCT_DT'].apply(lambda x: x.strftime("%Y"))
cln_df['FCT_MT_NUM']=cln_df['FCT_DT'].apply(lambda x: x.strftime("%m"))
cln_df['FCT_CD']=cln_df['FCT_YR_NUM']+'-'+cln_df['FCT_MT_NUM']
cln_df['VLM_UNT_CD'] = np.where(~cln_df['VLM_QTY'].isnull(),'TM','-')
cln_df['CIF_PRC_CRRNCY_CD'] = np.where(~cln_df['VLM_QTY'].isnull(),'$/TM','-')
cln_df['BDGT_CIF_PRC_CRRNCY_CD'] = np.where(~cln_df['VLM_QTY'].isnull(),'$/TM','-')
```

```python
cln_df['PPLTN_DT']=datetime.datetime.today().strftime("%m/%d/%Y")
```
